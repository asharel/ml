<H3>Obteniendo los datos del csv</H3>

```python
dataset = pd.read_csv('C:\\Users\Sara\ml\LAB1\Recursos\indian_liver_patient.csv',delimiter=',',header=0)
dataset
```

```python
# Header
header = []
for row in dataset:
    header.append(row)
header
```

<H3>Cambiando valores nominales</H3>

```python
dataset.replace("Male",0, True)
dataset.replace("Female",1,True)
dataset
```


<H3>Datos de estadisticos de las variables</H3>


```python
print(dataset.describe()) #Descripción de los datos
```

<H3>Normalizando variables</H3>

```python
dataset['Alkaline_Phosphotase'] = (dataset['Alkaline_Phosphotase'] - dataset['Alkaline_Phosphotase'].mean()) / (dataset['Alkaline_Phosphotase'].std())
dataset['Alamine_Aminotransferase'] = (dataset['Alamine_Aminotransferase'] - dataset['Alamine_Aminotransferase'].mean()) / (dataset['Alamine_Aminotransferase'].std())
dataset['Aspartate_Aminotransferase'] = (dataset['Aspartate_Aminotransferase'] - dataset['Aspartate_Aminotransferase'].mean()) / (dataset['Aspartate_Aminotransferase'].std())
dataset
```
